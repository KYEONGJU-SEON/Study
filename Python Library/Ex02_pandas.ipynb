### dictionary 구조를 활용하여 Series 생성하기
- 키와 밸류값을 가지는 데이터 구조 
- key는 index
- value는 value

```python
# 포켓몬이름 : 능력치
data = {'피카츄':9631,'파이리':5862,'꼬부기':1235,'이상해씨':2685}
pokemon = pd.Series(data)
```

```python
#레벨업한 능력치가 들어있는 pokemon_up
data2 = {'피카츄' : 10222,'파이리': 6524,'뮤':9999,'메타몽':2560}
pokemon_up = pd.Series(data2)
pokemon_up
```

```python
# 올라간 능력치 계산
stat = pokemon_up - pokemon
stat
#Nan(Not a number) = 결측치(비어있다)
```

```python
# 비어있지않은 데이터만 확인하기
stat.notnull() # 논리형 결과가 출력된다
# 비어있지않은 데이터만 보고싶다면? --> 불리언 인덱싱
stat[stat.notnull()]
# 비어있는 데이터만 확인하기!
stat[stat.isnull()]
```

```python
# 레벨업해서 올라간 능력치의 증감률(%) 구하기!
#(올라간데이터- 원래데이터)/원래데이터 *100
ch = (pokemon_up - pokemon) / pokemon *100
```

```python
# 비어있지 않은 데이터 출력
ch[ch.notnull()]
# 비어있는 데이터 출력
ch[ch.isnull()]
```

### 카테고리 만들기
- 데이터를 범주화 해주는 작업

```python
#나이데이터로 카테고리생성

#1. 데이터
ages = [0,2,10,32,20,23,61,100,54,16,38,40,88,73]
#2. 범주(bins)
bins= [-1,18,35,65,100]
#3. 카테고리 명칭
labels = ['미성년자','청년','장년','노년']
cate = pd.cut(ages,bins,labels = labels)
cate
```

- [ages 데이터들의 카테고리화된 각각의 결과들]
- 데이터의 개수
- 카테고리의 개수 , 카테고리 이름

```python
# 카테고리만 확인하고 싶을때
cate.categories
```

```python
# ages 데이터를 가지고 DataFrame 생성하기
age_df = pd.DataFrame(ages, columns=['ages'])
```

```python
# 카테고리 열 추가하기
age_df['카테고리']= cate
age_df
```

```python
#카테고리별로 데이터 개수 세는방법
age_df.count()
```

```python
data2019 = pd.read_csv('2019.csv',index_col='관서명',encoding ='EUC-KR')
data2020 = pd.read_csv('2020.csv',index_col='관서명',encoding ='EUC-KR')
data2021 = pd.read_csv('2021.csv',index_col='관서명',encoding ='EUC-KR')
```

```python
data2021
```

```python
#1. 특정연도에만 존재하는 경찰청이 존재 (drop)
data2021.drop('광주지방경찰청',inplace = True)
#df1.drop('2010',axis=1,inplace = True)
data2021
```

```python
#2. 연도별 범죄 데이터 총합 데이터 구하기('총합'이라는 컬럼 추가)
#score['합계'] = score.sum(axis=1)
data2021['2021총계'] = data2021.sum(axis=1)
data2021
```

```python
data2020['2020총계']=data2020.sum(axis=1)
data2020
```

```python
data2019['2019총계'] = data2019.sum(axis=1)
#data2019.loc[:,'살인':'폭력'] - 선생님
data2019
```

```python
# 3.구분(컬럼)의 데이터가 '발생건수'인 값들만 가져오기(구분==발생건수 불리언인덱싱)
data2019_1 =data2019[data2019['구분']=='발생건수']
data2019_1
```

```python
data2020_1 =data2020[data2020['구분']=='발생건수']
data2020_1
```

```python
data2021_1 =data2021[data2021['구분']=='발생건수']
```

```python
#4 총합을 기준으로 데이터 자르기 (인덱싱,슬라이싱) -> Series 데이터구조로 출력
data2019_2 = data2019_1['2019총계']
data2019_2
```

```python
data2020_2=data2020_1['2020총계']
```

```python
data2021_2 = data2021_1['2021총계']
```

```python
# 증감률 구하기
# (금년 - 작년) / 작년 * 100
def p (x,y):
    return (x-y)/y*100
```

```python
data2019_20 = p(data2020_2,data2019_2)
data2019_20.name='2019-2020 증감율'
```

```python
data2020_21 = p(data2021_2,data2020_2)
data2020_21.name='2020-2021 증감율'
```

```python
#6. 데이터프레임 병합 (총 5개 Series를 병합)
#19,19-20 증감률 , 20,20-21 증감률, 21
```

```python
pd.concat([data2019_2,data2019_20,data2020_2,data2020_21,data2021_2],axis=1)
```
