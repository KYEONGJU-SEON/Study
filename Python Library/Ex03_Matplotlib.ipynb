```python
#라이브러리 로딩하기
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Line Chart 그리기
x=[1,2,3,4]
y=[2,4,6,8]
plt.plot(x,y,ls = '--',lw=5, color = 'c',marker = 'o', mfc = 'r')

#속성값
#1. 라인의 스타일 설정가능 ls(LineStyle) --> 문자열로 기호를 적어주기! 
#2. 라인 두께 설정가능 lw(LineWidth) --> 그냥 숫자로 써주기~ 문자 xx
#3. 라인 색 변경가능 color -> marker와 line의 색 모두 변경
#4. 값의 위치 표시하기 marker
#5. marker 색 지정하기 mfc(Marker Face Color)

plt.grid()
#grid속성
# 모눈종이처럼 생긴 격자무늬 --> 그래프를 보기가 쉬워진다(show안쪽에 써야함 )

plt.show()
```

### 요일별 교통사고 차트 그리기 실습

```python
#csv 불러오기! 
data = pd.read_csv('Traffic_Accident_2017.csv',encoding ='EUC-KR')
data
```

```python
#dataFrame의 정보를 알려주는 기능 
data.info()
```

```python
# 상위 데이터만 확인하는 방법
data.head(n=5) # 상위 5개의 데이터만 출력해준다 
```

```python
#data에서 우리가 푤이한 '요일' 데이터만 가져오기 
data['요일'].count()
```

```python
# 각각의 요일별로 얼마나 발생했는지?
data2 = data['요일'].value_counts()
data2
```

```python
# 요일 정렬하기 월 화 수 목 금 토 일 
data3 = data2[['월','화','수','목','금','토','일']]
```

```python
# bar차트 그려주기 
#x <- 요일
#y <- 교통사고 발생건수

x= data3.index
y=data3.values

plt.bar(x,y,width = 0.3, color = 'c',edgecolor="y")

plt.xlabel('요일')
plt.ylabel('교통사고 \n 발생건수',labelpad = 50, loc = 'top').set_rotation(0)
plt.title('요일별 교통사고 \n 발생건수')

#그래프의 범위 지정(xlim, ylim)

plt.ylim(500,620)

plt.show()
```

```python
plt.barh(x,y)
```

```python
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]
# 시각화 하는 이유는 -> 한눈에 데이터를 보기 위함
plt.plot(x, label='x')
plt.plot(y, label='y')
plt.plot(z, label='z')
# 범례 설정하기 -> 참고사항
# plot 해줄때 label 이름 적어주기!
plt.legend()
# x축 이름 설정하기
plt.xlabel('x_label')
# y축 이름 설정하기
plt.ylabel('y_label')
# Chart 이름 설정하기
plt.title('차트')
plt.show()
```

```python
# 모든 컬럼을 출력하는 방법
pd.set_option('display.max_columns',None)
data
```

```python
# 사고유형 대분류에서 '차대차' 사건만 가져오기 --> 불리언인덱싱
data['사고유형_대분류'][data['사고유형_대분류']=='차대차']
```

```python
#총데이터에서 각각 얼마나 사고가 발생했는지
data['사고유형_대분류'].value_counts()
```

```python
acc = data[data['사고유형_대분류']=='차대차']
acc
```

```python
#우리가 필요한 사상자수와 발생지 시도 데이터만 가져오기
data1 = acc[['사상자수','발생지시도']]
data1
```

```python
#각각의 지역이 몇번 나와싸는지만 세어주는 결과값 - > 우리가 원하는 데이터 아님 
data1['발생지시도'].value_counts()
```

```python
#group by()
#같은 값을 하나로 묶어서 총계를 구해주고싶을때 사용 
result = acc[['사상자수','발생지시도']].groupby('발생지시도').sum()
# groupby()함수를 사용하면 기준 열이 자동으로 인덱스로 간다
# 만약 index로 사용하지않고싶다면? as_index = False로! 
result1 = result['사상자수']
result1
```

```python

x= result1.index
y= result1.values


# y축 이름 설정하기
plt.ylabel('사상자수',labelpad = 30).set_rotation(0)
# Chart 이름 설정하기
plt.title('차vs차 교통사고의 사상자수')
#plt.ylabel('교통사고 \n 발생건수',labelpad = 50, loc = 'top').set_rotation(0)

plt.bar(x,y)
plt.show()
```
