## 문자열 포매팅
- format 함수사용
- %기호 사용
- f문자열을 사용


```python
s = "오늘은 8월 1일입니다. 1일은 월요일 입니다."
print(s)
```

```python
s1 = "오늘은 8월 2일입니다. 2일은 화요일 입니다."
print(s1)
```

```python
date = 2
day = "화"
```

```python
s = '오늘은 8월 {}일 입니다.{}일은 {}요일 입니다.' .format(date,date,day)
print(s)
```

```python
s = '오늘은 8월 %d일입니다. %d일은 %s요일 입니다.'%(date,date,day)
print(s)
```

```python
s = f'오늘은 8월 {date}일입니다. {date}일은 {day}요일 입니다.'
print(s)
```

```python
x=100
y=200
sum2 = x+y
print('{}와 {}의 합은 {}입니다.'.format(x,y,sum2))
print( f'{x}와 {y}의 합은 {sum2}입니다')
print('%d와 %d의 합은 %s입니다.'%(x,y,sum2))
```

```python
num1 = 10
str1 = "7"
print(str(num1)+str1)
print(num1+int(str1))
```

```python
num1 = 23
num2 = 3
sum = num1 + num2
minus = num1-num2
mul = num1*num2
div = num1/num2

print('더하기 결과 : %d' %(sum))
print('빼기 결과 : %d' %(minus))
print('곱하기 결과 : %d' %(mul))
print('나누기 결과 :%f' %(div))

print('더하기 결과 : {}' .format(sum))
print(f'빼기 결과 : {minus}')
```
